```python
def get_coords(top_left, bottom_right):

    coord = [top_left[0], top_left[1]]
    lat = top_left[0]
    lon = top_left[1]
    coord = [lat, lon]
    coords = []
    while True:
        index = 0
        if lat > bottom_right[0]:
            lat = coord[0] - 0.08
        else:
            if lon < bottom_right[1]:
                lon = coord[1] + 0.08
                lat = top_left[0]
            else:
                break

        coord = [lat, lon]
        coords.append(coord)
    
    return coords
```
