```python
def get_columns_info(col_name):
    cols_list = inspector.get_columns(col_name)
    
    print("Table Name: {}".format(col_name))
    print("")
    
    for column in cols_list:
        print("Name: {} \t Type: {}".format(column['name'], column['type']))

get_columns_info('SURVEY')
```

```python
def get_columns_list(col_name):
    cols_list = inspector.get_columns(col_name)
    column_list = []
    print("Table Name: {}".format(col_name))
    print("")
    
    for column in cols_list:
        column_list.append(column['name'])
#         print("Name: {} \t Type: {}".format(column['name'], column['type']))
    return column_list
    
    
get_columns_list('BEGINEND')
```

```python
def dist_cal(lat1,long1,lat2,long2):
    l1 = np.radians(lat1)
    l2 = np.radians(lat2)
    lo1 = np.radians(long1)
    lo2 = np.radians(long2)

    dlat = l2-l1
    dlong = lo2-lo1
    ans = pow(sin(dlat / 2), 2) +cos(l1) * cos(l2) *pow(sin(dlong / 2), 2);

    ans = 2 * asin(sqrt(ans));
    r = 3958.8
    return(ans*r)
```
